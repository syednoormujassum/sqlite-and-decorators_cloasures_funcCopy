# Function Copy

```python
def add(a,b):
    c=a+b
    return c
```

```python
add(4,7)
```

```python
add1=add

# copying original function to a new instance
```

```python
add1(9,8)
```

```python
del add

# deleting original function
```

```python
add(3,4)
```

```python
add1(23,3)
```

- As you can see above when a function is copied over to a new instance and original function is deleted. Still the new instance works as original function.
