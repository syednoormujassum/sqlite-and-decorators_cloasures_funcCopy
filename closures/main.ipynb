# Closures

```python
def main(func):
    msg="Hey I am from Main function."
    def sub_main(func):
        print("Starting of Sub Main function")
        func(msg)
        print("Staring of second statement post execution of main function")
    return sub_main
```

```python
main(sub_main)
```

```python
@main
def sub_main2(func):
    print("Starting of Sub Main function")
    func(msg)
    print("Staring of second statement post execution of main function")
    return sub_main2
```

```python
sub_main2(print)
```

```python
import logging

logger=logging.getLogger("myLogger")
logging.basicConfig(level=logging.DEBUG)
def log_decorator(func):
    
    def wrapper(*args, **kwargs):
        
        try:
            logger.info(f"Logging started.{func.__name__}")
            func(*args,**kwargs)
            logger.info("Logging completed.")
        except Exception as e:
            logger.error(f"An error occurred: {e}")
    
    return wrapper
        
```

```python
@log_decorator
def add(a,b):
    print(a+b)
add(4,5)
    
```

```python
def simple_decorator(func):
    def wrapper():
        logger.info("Logging started.")
        print(">>> Starting function")
        
        func()
        print(">>> Function finished")
        logger.info("Logging complete.")
    return wrapper

@simple_decorator
def greet():
    print("Hello, World!")
greet()
```
